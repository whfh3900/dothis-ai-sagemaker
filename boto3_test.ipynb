```python
import os
import boto3
import json

# S3에 액세스할 수 있는 클라이언트 생성
s3 = boto3.client('s3')

# S3 버킷의 리스트 가져오기
response = s3.list_buckets()

# 버킷 리스트 출력
print('S3 버킷 리스트:')
for bucket in response['Buckets']:
    print(bucket['Name'])
```

```python
# S3에서 파일 가져오기
response = s3.get_object(Bucket='dothis-ai-test', Key='data/video_data_20230414.txt')

# 파일 내용 읽기
file_content = response['Body'].read().decode('utf-8')

# 파일 내용 출력
file_content[1:].split("\r\n")
```

```python
# S3에서 파일 가져오기
response = s3.get_object(Bucket=os.getenv("S3_BUCKET"), Key=os.getenv("HYPERPARAMETER_PATH"))

# 파일 내용 읽기
file_content = response['Body'].read().decode('utf-8')

# 파일 내용 출력
json.loads(file_content)

```
